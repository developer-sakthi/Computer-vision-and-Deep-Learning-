# Forward Propagation in a Single Neuron

## Aim

To apply forward propagation in a single neuron and analyze the effect of different activation functions on the output.

---

## Given

- **Input Vector:** `X = [1.0, 2.0]`
- **Weights:** `W = [0.5, -0.6]`
- **Bias:** `b = 0.1`

---

## Task 1: Manual Computation

### Step 1: Compute Linear Combination (Z)

**Formula:**

```text
Z = (X₁ × W₁) + (X₂ × W₂) + b
```

**Substituting the values:**

```text
Z = (1.0 × 0.5) + (2.0 × -0.6) + 0.1
Z = 0.5 - 1.2 + 0.1
Z = -0.6
```

**Linear Combination (Z):**

```text
Z = -0.6
```

---

### Step 2: Apply Sigmoid Activation

**Formula:**

```text
Sigmoid(Z) = 1 / (1 + e^(-Z))
```

**Substituting Z = -0.6:**

```text
Sigmoid(-0.6)
= 1 / (1 + e^(0.6))
≈ 1 / (1 + 1.8221)
≈ 0.3543
```

**Sigmoid Output:**

```text
≈ 0.354
```

---

### Step 3: Apply ReLU Activation

**Formula:**

```text
ReLU(Z) = max(0, Z)
```

**Substituting Z = -0.6:**

```text
ReLU(-0.6)
= max(0, -0.6)
= 0
```

**ReLU Output:**

```text
0
```

---

## Final Results

| Computation | Value |
|------------|--------|
| Linear Combination (Z) | `-0.6` |
| Sigmoid Output | `0.354` |
| ReLU Output | `0` |

---

## Task 2: Python Implementation

```python
import numpy as np

# Input vector, weights, and bias
X = np.array([1.0, 2.0])
W = np.array([0.5, -0.6])
b = 0.1

# Linear combination
Z = np.dot(X, W) + b

# Sigmoid activation
sigmoid = 1 / (1 + np.exp(-Z))

# ReLU activation
relu = max(0, Z)

# Display results
print("Z =", Z)
print("Sigmoid Output =", sigmoid)
print("ReLU Output =", relu)
```

### Expected Output

```text
Z = -0.6
Sigmoid Output = 0.35434369377420455
ReLU Output = 0
```

---

## Conclusion

- Forward propagation computes the weighted sum of inputs and bias to obtain the neuron input `Z`.
- The **Sigmoid** activation function maps the value to a range between `0` and `1`.
- The **ReLU** activation function outputs `0` for negative inputs.
- Different activation functions produce different outputs even when the same input `Z` is used.
