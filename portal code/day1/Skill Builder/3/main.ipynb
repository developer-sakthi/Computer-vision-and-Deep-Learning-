# Forward Propagation & ReLU Behavior in a Neuron

## Aim

To compute neuron outputs using forward propagation and observe how ReLU behaves for different input values.

---

## Given (Task 1)

- **Input Vector:** `X = [2.0, -1.0]`
- **Weights:** `W = [0.4, 0.6]`
- **Bias:** `b = -0.2`

---

# Task 1: Computation

## Step 1: Compute Linear Combination (Z₁)

### Formula:

```text
Z = (X₁ × W₁) + (X₂ × W₂) + b
```

### Substitution:

```text
Z₁ = (2.0 × 0.4) + (-1.0 × 0.6) + (-0.2)
```

```text
Z₁ = 0.8 - 0.6 - 0.2
```

```text
Z₁ = 0.0
```

### Linear Output:

```text
Z₁ = 0.0
```

---

## Step 2: ReLU Activation (Z₁)

### Formula:

```text
ReLU(Z) = max(0, Z)
```

### Calculation:

```text
ReLU(0.0) = max(0, 0.0) = 0
```

### Output:

```text
ReLU Output 1 = 0
```

---

# Task 2: Python Implementation

```python
import numpy as np

# Given values
X = np.array([2.0, -1.0])
W = np.array([0.4, 0.6])
b = -0.2

# Forward propagation
Z1 = np.dot(X, W) + b
relu1 = max(0, Z1)

print("Z1 =", Z1)
print("ReLU Output 1 =", relu1)
```

---

# Task 3: Computation (New Input)

## Given:

- **Input Vector:** `X = [-2.0, 1.0]`
- **Weights:** `W = [0.4, 0.6]`
- **Bias:** `b = -0.2`

---

## Step 1: Compute New Linear Combination (Z₂)

### Substitution:

```text
Z₂ = (-2.0 × 0.4) + (1.0 × 0.6) + (-0.2)
```

```text
Z₂ = -0.8 + 0.6 - 0.2
```

```text
Z₂ = -0.4
```

### Linear Output:

```text
Z₂ = -0.4
```

---

## Step 2: ReLU Activation (Z₂)

```text
ReLU(Z₂) = max(0, -0.4) = 0
```

### Output:

```text
ReLU Output 2 = 0
```

---

# Final Results

| Case | Z Value | ReLU Output |
|------|--------|-------------|
| Input 1 `[2.0, -1.0]` | `0.0` | `0` |
| Input 2 `[-2.0, 1.0]` | `-0.4` | `0` |

---

# Conclusion

- Forward propagation computes a weighted sum of inputs plus bias.
- ReLU activation outputs:
  - Positive values → unchanged
  - Zero or negative values → 0
- This shows why ReLU introduces sparsity in neural networks.
