# Debugging Sigmoid Activation Function in Neural Networks

## Aim

To identify and correct an error in the Sigmoid activation function and verify the correct output.

---

## Task 1: Debugging

### Given Incorrect Function

```python
def sigmoid(z):
    return 1 / (1 + np.exp(z))
```

---

### ❌ Identify the Mistake

The error is in the exponent sign.

- Correct Sigmoid formula is:

```text
σ(z) = 1 / (1 + e^(-z))
```

- But the given implementation uses:

```text
np.exp(z)   ❌ (wrong sign)
```

This produces incorrect results because it grows instead of decaying.

---

### ✅ Correct Implementation

```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

---

## Task 2: Output Check

### Given

```text
Z = -1
```

---

## Step 1: Wrong Sigmoid Output

Using incorrect function:

```text
sigmoid(z) = 1 / (1 + e^(z))
```

Substitute `z = -1`:

```text
= 1 / (1 + e^(-1))
= 1 / (1 + 0.3679)
= 1 / 1.3679
≈ 0.731
```

### ❌ Wrong Output:
```text
≈ 0.731
```

---

## Step 2: Correct Sigmoid Output

Using correct function:

```text
sigmoid(z) = 1 / (1 + e^(-z))
```

Substitute `z = -1`:

```text
= 1 / (1 + e^(1))
= 1 / (1 + 2.7183)
= 1 / 3.7183
≈ 0.269
```

### ✅ Correct Output:
```text
≈ 0.269
```

---

## Final Results

| Case | Output |
|------|--------|
| Wrong Sigmoid | `0.731` |
| Correct Sigmoid | `0.269` |

## Python Verification Code

```python
import numpy as np

Z = -1

# Wrong implementation
wrong_sigmoid = 1 / (1 + np.exp(Z))

# Correct implementation
correct_sigmoid = 1 / (1 + np.exp(-Z))

print("Z =", Z)
print("Wrong Sigmoid Output =", wrong_sigmoid)
print("Correct Sigmoid Output =", correct_sigmoid)
```
