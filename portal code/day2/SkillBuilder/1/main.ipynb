```python
import math

y_true = 1
y_pred = 0.68

# Mean Squared Error formula: (y_true - y_pred)^2
mse = (y_true - y_pred) ** 2

# Binary Cross Entropy formula: -[y_true * ln(y_pred) + (1 - y_true) * ln(1 - y_pred)]
bce = -(y_true * math.log(y_pred) + (1 - y_true) * math.log(1 - y_pred))


print(f"MSE: {mse:.4f}")  # Output: 0.1024
print(f"BCE: {bce:.4f}")  # Output: 0.3857

```
