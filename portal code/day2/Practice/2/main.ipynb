```python
import numpy as np

# Given values
y = 1
y_hat = 0.6

# MSE
mse = (y - y_hat) ** 2

# Binary Cross Entropy
bce = -(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))

# Weight update parameters
w = 0.8
dw = 0.5

# Learning rates
eta1 = 0.1
eta2 = 0.5

# Updated weights
w_new1 = w - eta1 * dw
w_new2 = w - eta2 * dw

# Display results
print("MSE =", mse)
print("Binary Cross Entropy =", round(bce, 4))
print("Updated weight (η = 0.1) =", w_new1)
print("Updated weight (η = 0.5) =", w_new2)
```
