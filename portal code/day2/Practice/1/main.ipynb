```python
import numpy as np

# Given values
y = 0
y_hat = 0.3
x = 1.5
w = 0.6
eta = 0.2

# BCE gradient with respect to prediction
dL_dyhat = -(y / y_hat) + ((1 - y) / (1 - y_hat))

# Gradient of prediction with respect to weight
dyhat_dw = x

# Chain rule
dL_dw = dL_dyhat * dyhat_dw

# Weight update
w_new = w - eta * dL_dw

# Display results
print("dL/dŷ =", round(dL_dyhat, 4))
print("dŷ/dw =", dyhat_dw)
print("dL/dw =", round(dL_dw, 4))
print("Updated weight =", round(w_new, 4))
```
